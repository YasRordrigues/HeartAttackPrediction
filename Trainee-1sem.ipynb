```python
list_Idades = [i for i in df['age']]
contagem = Counter(list_Idades)

x = [i for i in contagem.keys()]
y = [i for i in contagem.values()]
print(x,y)
plt.barh(x,y)
plt.title('Idades Pesquisadas')
plt.xlabel('quantidade por idade')
plt.ylabel('idades')
plt.show()
```

```python
chest_pain = [i for i in df['cp']]

conta_cp = Counter(chest_pain)
print(conta_cp)

cp_types = [i for i in conta_cp.keys()]
cp_contagem = [i for i in conta_cp.values()]
print(cp_types,cp_contagem)

plt.bar(cp_types,cp_contagem)
plt.title('dores no peito')
plt.xlabel('classificaÃ§Ã£o')
plt.ylabel('quantidade')
plt.show()
```
